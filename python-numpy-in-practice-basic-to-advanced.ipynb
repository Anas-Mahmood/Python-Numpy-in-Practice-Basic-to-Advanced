```python
arr.sort() #Sorts arr
arr
```

```python
arr.sort(axis=0) #Sorts specific axis of arr
arr
```

```python
trans=np.array([(1,2,3),(4,5,6)])
trans.T #Transposes arr (rows become columns and vice versa)
```
